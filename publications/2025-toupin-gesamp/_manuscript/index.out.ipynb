# Abstract

Environmental scientific advisory organizations and expert groups play a crucial role in synthesizing and disseminating knowledge to various stakeholders to inform practice and policy. This work is commonly disseminated through grey literature (e.g., reports and policy briefs). However, assessing the visibility, use, or impact of these outputs can be challenging because they often lack standard identifiers, such as DOIs, and are therefore not indexed in citation indexes and other databases commonly used to measure engagement using indicators like the number of views, downloads, or citations. Using the reports produced by the Joint Group of Experts on the Scientific Aspects of Marine Environmental Protection (GESAMP) as a case study, this study explores the use of various data sources (Web of Science, Overton and news indexing platforms) to track references to GESAMP reports in the scholarly literature, policy documents, and news media. Our findings indicate that attention to GESAMP reports has increased over the past twenty years, particularly in the last decade. GESAMP reports appear to generate interest in research fields dealing with environmental pollution and ocean management. Outside of academia, while sponsoring organizations tend to cite GESAMP the most, the reports also attract the interest of national and regional environmental organizations. Our study provides empirical evidence on how such reports and the organizations producing them can reach different audiences.

# Introduction

The production and dissemination of scientific knowledge constitute a critical foundation for the development of evidence-based policy frameworks and coordinated societal responses to pressing socioenvironmental challenges. In the context of accelerating climate change, ecological degradation, and widening social inequities, integrating robust empirical research into decision-making processes is essential for ensuring that interventions are both effective and equitable. Scientific inquiry not only elucidates the underlying drivers and systemic interdependencies of environmental phenomena but also enables the modelling of future scenarios and the evaluation of policy outcomes across temporal and spatial scales ([Government of Canada, 2020](#ref-canada2020climatescience)). The dissemination and engagement with scientific knowledge facilitate its translation into action ([Tangney, 2022](#ref-tangney2022)). There is an imperative to produce knowledge that responds to policy priorities. As such, the strategic creation and circulation of scientific knowledge must be recognized as a cornerstone of informed governance and collaborative action in the Anthropocene ([Morgan & Di Giulio, 2018](#ref-morgan2018)).

International scientific advisory bodies, such as the United Nationsâ (UN) Joint Group of Experts on the Scientific Aspects of Marine Environmental Protection (GESAMP), bridge the scientific and public spheres by conducting research, synthesizing knowledge, and formulating recommendations for policymakers on key issues, including marine pollution and ocean acidification. Established in 1969, GESAMP functions under the auspices of ten UN organizations with overlapping maritime and ocean interests. Studies and assessments are conducted by dedicated Working Groups, comprised of a multidisciplinary group of experts on the subject matters of specific projects to provide insights alongside advice for science-led solutions.Â 

GESAMP primarily contributes knowledge about ocean sustainability issues to the scientific community and worldwide marine protection advisory bodies through knowledge reports and thematic documents. GESAMP has produced over 120 sessional and topical reports, ranging from specific issues relating to marine protectionâsuch as their 1985 study Atmospheric Transport of Contaminants into the Mediterranean Regionâto higher-level publications that survey the current state of our oceans, such as A Sea of Troubles: Issues in Focus (2001), and more recent assessments like Marine Biofouling: Non-Indigenous Species and Management Across Sectors (2024), which explores the ecological and regulatory challenges posed by biofouling and its role in the spread of invasive species.

GESAMPâs research has been characterized as ranging âfrom in-depth reviews of critical marine processes, perturbations and contaminant categories to advice on monitoring, assessment and modelling, and guidance on strategic aspects of marine environmental protection and managementâ ([Wells et al., 2002, p. 80](#ref-wells2002caring)). Indeed, GESAMP is unique in that it is made up of individuals from autonomous organizations, ultimately legitimizing its activities and research credibility for the international community as it operates âat armâs length from a government, although in this case, the government is not one body but is itself a network of officials drawn from eight \[now ten\] governing bodiesâ ([McLaren, 1990, p. 657](#ref-mclaren1990network)).

Studies on communicating knowledge to decision-making bodies ([Cossarini et al., 2014](#ref-cossarini2014communicating)) , stakeholders ([Soomai et al., 2013](#ref-soomai2013communicating)), and informing policies ([Lawrence et al., 2014](#ref-lawrence2014evidence)) provide context for understanding how information relating to coastal and marine protection produced by expert advisory groups is translated into the public sphere, used to shape environmental management decisions at all levels, and communicated to stakeholders who can be informed by the knowledge produced by these groups.Â Â 

Thus, contemporary scientific research and policymaking have built upon knowledge produced by GESAMP over the decades ([Watson-Wright et al., 2024](#ref-watsonwright2024gesamp)). Twenty years ago, studies by Cordes ([2003](#ref-cordes2003grey)) and MacDonald et al. ([2004](#ref-macdonald2004grey)) specifically aimed to capture the engagement with and use of GESAMP reports in the scholarly and policy literature. Our study serves as a follow-up to this work, incorporating recent reports and integrating new indicators, such as mentions of the reports in policy documents and news media.

## Challenges in measuring the use and impact of grey literature

Grey literature is a term used to encompass a wide range of documents produced all levels and sectors of government, academia, industry, or other ([SchÃ¶pfel, 2011](#ref-schopfel2011prague)). Grey literature does not usually go through the scientific peer review process, and it can thus be published and disseminated quickly and communicate complex information to non-specialist audiences in an accessible format ([Pappas & Williams, 2011](#ref-pappas2011grey)).

For these reasons, its inclusion in systematic reviews and use in policymaking has risen over the past few decades ([Adams et al., 2017](#ref-adams2017); [Babarczy et al., 2024](#ref-babarczy_national_2024); [Lawrence et al., 2014](#ref-lawrence2014evidence); [MacDonald et al., 2016](#ref-macdonald2016science); [Paez, 2017](#ref-paez2017); [Pappas & Williams, 2011](#ref-pappas2011grey); [Thatje et al., 2007](#ref-thatje2007elNino)). However, its lack of indexing in citation databases makes assessing its use and impact difficult, which is important for encouraging successful types of grey literature and recognizing their contributions to knowledge advancement ([Bickley et al., 2020](#ref-bickley2020impact)). Wilkinson et al. ([2014](#ref-wilkinson2014substance)) proposed a web impact report indicator to identify web pages that cite or mention the documents in question, to be used in conjunction with other altmetric indicators (e.g., newspaper mentions). While such an indicator might account for use, the authors note that its limitations endure in assessing impact. It remains a managerial decision to gauge whether such indicator results are significant ([Wilkinson et al., 2014](#ref-wilkinson2014substance)). Bickley et al. ([2020](#ref-bickley2020impact)) used a semi-automated method, using Scopus (API) cited reference searches to locate grey literature citations and the Google Books API to identify monograph citations. They advanced a new approach in 2022 by using Publish or Perish and Webometric Analyst software and found it possible to match references in grey literature to academic publications (Bickley et al., 2022). Saleem et al. ([2023](#ref-saleem2023referencing)) investigated evidence produced and relied on by non-academic organizations in the policy sphere and found an increasing reliance on grey literature.Â 

Attempts have been made to assess the use and impact of GESAMP-specific literature ([Cordes, 2003](#ref-cordes2003grey); [Hutton, 2009](#ref-hutton2009scientific); [MacDonald et al., 2004](#ref-macdonald2004grey)). Cordes ([2003](#ref-cordes2003grey)) conducted a citation analysis using the Web of Science, and MacDonald et al.Â (2004) employed a similar approach to assess the use of GESAMP literature.

Hutton ([2009](#ref-hutton2009scientific)) utilized the Web of Science, Google, and Google Scholar to investigate the influence of grey literature, advocating for a multifaceted approach to understanding GESAMPâs influence through diverse citation data and indicators. Adie ([2014](#ref-adie2014grey)) similarly suggested that investigations of the use and impact of grey literature could be improved through the assignment of persistent identifiers, advanced metadata standards, and open, central indexes or repositories of grey literature.

Examining a range of altmetric indicators, in addition to citations, and utilizing new data sources, such as the Overton[1] database (founded by Adie), provides a means of capturing the impact of grey literature within and beyond the scientific community.

# Research Objectives

This study examines engagement with GESAMP reports in research, policy, and the news media. As scholarly citations, policy mentions, and news mentions may be associated with different forms of scientific and public contributions to knowledge and policy, they illuminate attention given toward the outputs of scientific organizations ([Afinowi & Nhamo, 2025](#ref-afinowi2025mapping); [Alperin et al., 2024](#ref-alperin2024second); [Haustein et al., 2016](#ref-haustein2016interpreting); [Toupin et al., 2023](#ref-toupin2023shark)). This study investigates the presence of GESAMPâs research outputs in the scholarly and public spheres. It aims to answer the following research questions:

**RQ1.** What is the prevalence of GESAMP publications in the scholarly literature, policy documents, and the news?

**RQ2.** What are the characteristics of GESAMPâs most cited and mentioned publications?

Since the reports produced by GESAMP have employed different formats and addressed various topics and issues over the years, engagement may be variable in research, policy, and news media, as well as being embedded in specific contexts of production and communication. Also significant to this study is an examination of the types of attention received by GESAMP publications. This study thus assesses GESAMPâs presence in the scholarly literature and public sphere as indicated by the number of citations, news mentions, and policy mentions of their reports.

## Policy Citations

GESAMP thematic reports, annual reports, histories, proceedings, and executive summaries were searched in the Overton database to collect citations by policy documents. Overton is the largest searchable index of policy documents, guidelines, think tank publications and working papers. Overton relies on Crossref, OpenAlex, and Wikidata infrastructures to collect citations and mentions of these documents. The Overton database provides citation contexts by indicating where a scholarly output is used in the text of a policy document (when a bibliography section is provided in a policy document, it maps to its DOI) (Szomszor & Adie, 2023). As the referencing practices of grey literature can vary, Overton typically relies on a set of scored features to capture citations. If the overall score exceeds a certain threshold, the citation context is matched to Crossref to capture citations to scholarly literature or the rest of the Overton database for policy documents. Overton also provides data such as authors of policy documents, source type (e.g., government, think tank), document topics, associated Sustainable Development Goals and the name of the source organization that produced the policy document (e.g., UNESCO, Institute for European Environmental Policy, Government of Canada) as matched by their internal methods.

GESAMP literature was searched individually by title. A search for GESAMP was also run, supplementing missing titles indexed under different names. Most (118 out of 120) of GESAMPâs outputs were found in Overton, resulting in 2,112 citations. After removing self-citations between GESAMP publications, we retrieved 742 references in 551 distinct publications by 103 organizations linking to GESAMP publications. Notably, many policy citations are self-references by the organization to its own literature. However, this may be due to an artefact in the data collection process as many of these citations referred to the list of GESAMPâs previous reports present at the end of each of its publications.Â 

For the analysis, we counted the number of citations per year, the number of citations per citing organization, and the number of citations to each report.

A search for âGESAMPâ or âgroup of experts on the scientific aspects of marineâ retrieved 189 results, which were compiled into a CSV file.Â

We subsequently examined the context in which GESAMP was mentioned to identify the specific report that was mentioned. The resulting dataset included 217 news items related to GESAMP publications.Â 

For the analysis, we counted the number of news mentions to GESAMP per year, the number of mentions per media source, and the number of news mentions to individual reports.

# Discussion and conclusion

This study shows that since Cordes (2003) and MacDonald et al.Â (2004), there has been a steady increase in the number of citations to GESAMP publications in scholarly literature and a sharp increase since 2016. This trend is noticeable in news and policy documents, with mentions increasing steadily from 2000 onwards. These trends may be due to increased public attention to climate change and ocean pollution. For instance, the UNâs First World Ocean Assessment, released in 2017, exposed the degraded state of the oceans (UNESCO, 2017) and catalyzed national action. Canada, for example, launched a \$1.5 billion Oceans Protection Plan (OPP), the largest federal government investment to protect oceans (DFO Maritimes Region OPP Atlantic Hub, 2020). Such efforts around the globe may have spurred research drawing on GESAMP publications. As 2021 marked the beginning of the UN Decade of Ocean Science for Sustainable Development, GESAMP reports will likely continue to be needed and used, as evidenced by a rise in news mentions in 2022.Â Â 

Scientific attention, defined as the number of citations, can be observed mainly in fields and journals dealing with environmental and marine pollution, such as Marine Pollution Bulletin, Science of the Total Environment, or Environmental Pollution, as they tend to cite GESAMP the most. However, Ocean & Coastal Management and Marine Policy are also among the top citing journals, showing environmental management research is another field interested in GESAMPâs work. Moreover, mentions in policy documents indicate the relevance of GESAMP for governance and official bodies. Though itâs mostly UN organizations that draw from GESAMP, the expert group is also referenced in documents produced by the Canadian and United Kingdom governments and by organizations such as the International Union for Conservation of Nature.

The prevalence of attention to Sources, fate, and effects of microplastics in the marine environment (Parts 1 and 2) reifies the importance of plastic pollution, especially microplastics, particularly in recent years. As a result of the degradation of plastic components in ocean waters, microplastics pose several threats to marine life and human health through their ingestion and accumulation in the food chain. These reports were highly discussed and mobilized in news media when published. In this regard, GESAMP work has been foundational in estimating the size of the problem and building a better understanding of the issue. This also instantiates how scientific advisory bodies serve a key role in discovering and building knowledge on new public issues by mobilizing resources at a larger scale. Overall, pollution, especially plastic, oil, or chemicals, is central to GESAMPâs most cited and mentioned work, thus highlighting its position as a key actor in tackling these issues. We also observe that GESAMPâs most mentioned work in policy documents and news media was published in recent years, a stark comparison to citations. This is likely the result of two phenomena: 1) the comparatively quicker pace at which public attention may accumulate to alternative formats and 2) digital transformations that make relevant information more easily accessible, especially to policymakers.Â 

These results reveal the global use of GESAMP publications, their relevance to specific locales, and their utility to organizations with diverse purposes not solely limited to environmental or marine protection. Nevertheless, GESAMP publications remain most cited by its sponsoring organizations, such as UNEP, the FAO, and the WMO. Adie (2023) notes that concerning policy, governments tend to mostly cite locally. In this way, international bodies such as the UN, citing GESAMP literature, reveal how its findings may apply to work and policy with a global focus and less to regionalized policy. GESAMP reports are nevertheless regularly mentioned by national organizations, as evidenced by governmental and NGO documents from Australia, Canada, New Zealand, and Finland, among others.Â 

This analysis of literature citing GESAMP publications used references in scholarly articles from the Web of Science, policy documents from Overton, and news mentions from Factiva, EurÃªka, ProQuest News and Newspapers, NexisUni, and Research Library. It finds that most citations are in scientific articles (4,713 citations from 4,082 distinct articles), followed by news sources (708 news mentions from 329 different news sources) and policy documents (742 policy citations from 551 publications published by 103 organizations linking to 118 GESAMP publications). It observes continuous growth of engagement with GESAMP publications over time, both in citations and in mentions of GESAMP literature in news media.Â 

This studyâs approach possesses several limitations. A search for GESAMP publications was conducted across databases by searching for âgesampâ and its full name. Reports were not searched for at an individual level, potentially limiting the comprehensiveness of retrieved results from databases. As exhaustive as our data sources may be, this report may present an undervaluation of GESAMP literature; Overton and Web of Science do not index full coverage of GESAMP publications. Further, availability bias in Overton and policy documents online may skew our dataset. Some countries have fewer policies available online than others or use references in the same ways, potentially excluding mentions of GESAMP. Regarding news sources, databases do not typically distinguish between media, newswire, or news-feeding services. Future work will be needed to disambiguate the provenance and different layers of news services in the data.
